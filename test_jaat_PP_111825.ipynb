Does not look like a title...

It probably means:

*   '13-1071.00':
SOC / ONET occupation code.
Code 15-1243.00 corresponds to Human Resources Specialist in ONET.

*   0.872:
Semantic match score between your job title string and the SOC code’s canonical title text.

*   -0.9:
Output of their title_value regression model (some internal “value” score, not a title).

*   'none':
Output of the multi-label title_feature model: which title features are present. In this example, it predicted no special features.

Let me try using the data package creators used in their deme. Did they extract job title from text? Or used already extracted title?

```python
from operator import itemgetter


data = pd.read_excel("https://raw.githubusercontent.com/Job-Ad-Research-at-QSB-LUC/JAAT/main/data/demo/JobAdsData2022_OSF.xlsx")
data['text'] = data['text'].str.strip()
data.text = data.text.replace(r'\n','. ', regex=True)
data.text = data.text.apply(lambda x: x.lower() if isinstance(x, str) else x)
# We will work with the columns 'job_title' and 'text' (a concatenation of all text)
data.head(1)
```

```python
data.job_title.to_list()
```

```python
data['TitleMatch'] = tt.get_title(data.job_title.to_list())
data['TM_onet_id'] = data.TitleMatch.apply(itemgetter(1)).astype(str)
```

### Conclusion 1

TitleMatch from JAAT does not extract job title from job description. It just turns it to ONET code, you need to extract job title yourself or have it in the data as a column to begin with.

According to the demo:
*   TitleMatch - Match Job Titles to O-NET Occupation Codes.
*   FirmNER - Extracts Candidate Firm Names from Job Ad (or other) Text.
*   TaskMatch - Match Job Ad Text to O-NET Task IDs
*   CREAM - Develop custom classifiers with embeddings by augmenting a human curated list of 'rules'. Demo uses Education Requirements.
